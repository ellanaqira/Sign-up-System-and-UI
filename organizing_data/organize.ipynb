    Read csv file and show the data frame

```python
# importing dependencies
import pandas as pd
```

```python
# read user_data.csv
user_df = pd.read_csv("template_data.csv")
# show the data frame
user_df
```

    - Find specific users by Gender

- Male Gender

```python
# find specific data (Gender = male)
male_user_df = user_df[user_df['Gender'] == "Male"]
male_user_df
```

```python
# find the number of user with Male gender
print(f"Data Frame size : {male_user_df.shape}")
print(f"Number of Male users : {male_user_df.shape[0]}")
```

- Female Gender

```python
# find specific data (Gender = female)
female_user_df = user_df[user_df['Gender'] == "Female"]
female_user_df
```

```python
# find the number of Female user
print(f"Data Frame size : {female_user_df.shape}")
print(f"Number Female users : {female_user_df.shape[0]}")
```

    Find specific users by year of birth

- Male users that born < 2000-12-31

```python
# find male users with born year < 2000-12-31
under_twoth_male = male_user_df[male_user_df['Date of birth'] < '2000-12-31']
print("Male user that born under < 2000-12-31\n")
print(under_twoth_male)
# Number of male users that born < 2000-12-31
print("="*75)
print(f"\nTotal : {under_twoth_male.shape[0]}")
```

- Male users that born >= 2000-12-31

```python
# find male users that born >= 2000-12-31
above_twoth_male = male_user_df[male_user_df['Date of birth'] >= '2000-12-31']
print("Male user that born under >= 2000-12-31\n")
print(above_twoth_male)
# Number of male users that born >= 2000-12-31
print("="*75)
print(f"\nTotal : {above_twoth_male.shape[0]}")
```

- Female users that born < 2000-12-31

```python
# find female users that born < 2000-12-31
under_twoth_female = female_user_df[female_user_df['Date of birth'] < '2000-12-31']
print("Female user that born under < 2000-12-31\n")
print(under_twoth_female)
# Number of female users that born < 2000-12-31
print("="*75)
print(f"\nTotal : {under_twoth_female.shape[0]}")
```

- Female users that born >= 2000-12-31

```python
# find female users that born >= 2000-12-31
above_twoth_female = female_user_df[female_user_df['Date of birth'] >= '2000-12-31']
print("Female user that born under >= 2000-12-31\n")
print(above_twoth_female)
# Number of female users that born >= 2000-12-31
print("="*75)
print(f"\nTotal : {above_twoth_female.shape[0]}")
```
